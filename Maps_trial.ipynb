```python
prueba = pd.DataFrame(eval(df_tube['close_stations'].iloc[0]), columns = ['name', 'lat', 'long'])
prueba.drop_duplicates('name',inplace = True)
prueba['lat_dif'] = prueba['lat'] - df_tube['latitude'].iloc[0]
prueba['long_dif'] = prueba['long'] - df_tube['longitude'].iloc[0]
prueba['sqrt'] = ((prueba['lat_dif']**2) + (prueba['long_dif']**2))**0.5
prueba.sort_values('sqrt', inplace = True)
prueba.reset_index(drop = True, inplace = True)
prueba
```
